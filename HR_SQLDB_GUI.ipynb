```python
#importing libraries for simple gui 
import PySimpleGUI as sg
from mysql import connector

#applying theme for the window

sg.theme("BluePurple")

#connecting to the database

def sqlconnect(pas):
    con = connector.connect(
    host = "localhost",
    user = "root",
    password = pas)
    return con

#defining buttond for different operations

l1 = [ [sg.Text("Welcome to Database Application",justification = "center",size = (50,1))],[sg.Text("")],
    [sg.Button("Password to connect to database",size=(20,2))],
      [sg.Text("",key="u",size = (20,1))],
     [sg.Button("Run User Query")],[sg.Text("_"*60)],
      [sg.Text("HR database",size = (50,1), justification = "center")],
      [sg.Button("Employees",size=(10,2)),sg.Button("Employees in various departments",size =(15,2)),
      sg.Button("Job Roles", size = (10,2)),sg.Button("Jobs and Salaries",size = (10,2))]
     ]

#defining size and layout for the window

l2 = [[sg.Text("Output",size = (10,10)),sg.Text("",size = (40,10),key = "out")]]

layout = [[sg.Column(l1)],
          [sg.Column(l2)]]

#giving name of window as database

window = sg.Window("database",layout)

while True:
    event, values = window.read()
    
    if event in (sg.WIN_CLOSED, 'Exit'):
        break
    if event == sg.WIN_CLOSED or event == 'Exit':
        break
        
    if event:
        
        if event == "Password to connect to database":
            password = sg.popup_get_text("password", password_char="*")
            sqlconnection = sqlconnect(password)
            window["u"].Update("Connected!")
            cur = sqlconnection.cursor()
#if the password entered by user is true then the operation continuous
        if event == "Run User Query":
            q = sg.popup_get_text("enter your query")
            cur.execute(q)
                
            for x in cur:
                window["out"].Update(list(x))
#if user selected employees then give output from database      
        if event == "Employees":
            
            query1 = "use HRdatabase;"
            query2 = "select count(employee_id) from employees;"
            
            cur.execute(query1)
            cur.execute(query2)
            
            for x in cur:
                window["out"].Update(list(x))
#employees working in various departments in the company       
        if event == "Employees in various departments":
            query1 = "use HRdatabase;"
            query2 = "select d.department_name, count(d.DEPARTMENT_NAME) from departments d inner join employees e on e.DEPARTMENT_ID = d.DEPARTMENT_ID group by d.DEPARTMENT_NAME;"
        
            cur.execute(query1)
            cur.execute(query2)
            
            results = []
            for r in cur:
                results.append(list(r))
            
            window["out"].Update(results)
# job roles in the company        
        if event == "Job Roles":
            query1 = "use HRdatabase;"
            query2 = "select j.JOB_TITLE, count(j.JOB_TITLE) from jobs j inner join employees e on j.JOB_ID = e.JOB_ID group by e.JOB_ID;"
            cur.execute(query1)
            cur.execute(query2)
            
            results = []
            for r in cur:
                results.append(list(r))
            
            window["out"].Update(results)
#pay roles according to the job        
        if event == "Jobs and Salaries":
            query1 = "use HRdatabase;"
            query2 = "select concat(e.first_name,' ',e.last_name),j.JOB_TITLE, max(e.salary) from jobs j inner join employees e on j.JOB_ID = e.JOB_ID group by e.JOB_ID order by max(e.salary) desc limit 5;"
            cur.execute(query1)
            cur.execute(query2)
            
            results = []
            for r in cur:
                results.append(list(r))
            
            window["out"].Update(results)
            
            
window.close()
```
